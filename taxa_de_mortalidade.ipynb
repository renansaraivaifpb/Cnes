```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/datasus/óbitos fetais/2.csv")
df = df.replace("-",0)
# Renomear a coluna 'Nome' para 'NomeCompleto'
df.rename(columns={'Sint sinais e achad anorm,exc morte súbita infânci': 'Sint sinais e achad anorm'}, inplace=True)
df.rename(columns={'Doenças do aparelho circulatório, mal definidas': 'Doenças do aparelho circulatório'}, inplace=True)
df.rename(columns={'Doenças do aparelho respiratório, mal definidas': 'Doenças do aparelho respiratório'}, inplace=True)
df.rename(columns={'Afecções originad período perinatal, mal definidas': 'Afecções originad período perinatal'}, inplace=True)
# Remover a última coluna
df = df.drop(df.columns[-1], axis=1)
# Remover a primeira linha
df = df.drop(0, axis=0)
# Redefinir os índices após a remoção
df.reset_index(drop=True, inplace=True)
# Transformar as colunas para o tipo int64
df['Sint sinais e achad anorm'] = df['Sint sinais e achad anorm'].astype(int)
df['Doenças do aparelho circulatório'] = df['Doenças do aparelho circulatório'].astype(int)
df['Doenças do aparelho respiratório'] = df['Doenças do aparelho respiratório'].astype(int)
df['Afecções originad período perinatal'] = df['Afecções originad período perinatal'].astype(int)

# Verificar os tipos de dados das colunas após a transformação
```

```python
# Fazer a interseção com base na coluna "Município"
resultado = pd.merge(df, obitos, on='Município', how='inner')

# O resultado conterá apenas as linhas com Município em comum entre os dois DataFrames
(resultado)
```

```python
import pandas as pd
import numpy as np

# Suponha que você tenha o DataFrame resultado com as colunas "Município", "2013_x", "2014_x", ..., "2021_y"

# Função para realizar a divisão com verificação de divisão por zero
def safe_divide(y, x):
    if x == 0:
        return np.nan
    else:
        return y / x

# Converter as colunas de ano para números
for ano in range(2013, 2022):
    col_x = f'{ano}_x'
    col_y = f'{ano}_y'
    resultado[col_x] = resultado[col_x].astype(float)
    resultado[col_y] = resultado[col_y].astype(float)

# Calculando as divisões
resultado['Município'] = resultado['Município']
for ano in range(2013, 2022):
    col_x = f'{ano}_x'
    col_y = f'{ano}_y'
    resultado[f'Divisão_{ano}'] = resultado.apply(lambda row: safe_divide(row[col_y], row[col_x]), axis=1)

# Criando um novo DataFrame com as colunas desejadas
novo_df = resultado[['Município'] + [f'Divisão_{ano}' for ano in range(2013, 2022)]]

# Exibindo o novo DataFrame com as divisões
(novo_df)
```

```python
# Substituir valores NaN por 0
novo_df = novo_df.fillna(0)
novo_df
```

```python
for ano in range(2013, 2022):
    coluna_divisao = f'Divisão_{ano}'
    novo_df[coluna_divisao] = novo_df[coluna_divisao] * 1000

# Exibindo o novo DataFrame com as divisões multiplicadas por 1000
(novo_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponha que você já tenha o DataFrame novo_df

# Selecionar apenas as colunas Divisão_2013, Divisão_2014, ..., Divisão_2021
colunas_divisao = [f'Divisão_{ano}' for ano in range(2013, 2022)]

# Calcular o desvio padrão (std) e o valor máximo (max) ao longo do tempo
std = novo_df[colunas_divisao].std()
maximo = novo_df[colunas_divisao].max()

# Criar um gráfico de linha para o desvio padrão (std) ao longo do tempo
plt.figure(figsize=(12, 6))
plt.plot(colunas_divisao, std, marker='o', label='Desvio Padrão (std)', color='b')
plt.title('Evolução do Desvio Padrão (std) ao Longo do Tempo')
plt.xlabel('Ano')
plt.ylabel('Desvio Padrão')
plt.xticks(rotation=45)
plt.legend()

# Criar um segundo gráfico de linha para o valor máximo (max) ao longo do tempo
plt.figure(figsize=(12, 6))
plt.plot(colunas_divisao, maximo, marker='o', label='Valor Máximo (max)', color='r')
plt.title('Evolução do Valor Máximo (max) ao Longo do Tempo')
plt.xlabel('Ano')
plt.ylabel('Valor Máximo')
plt.xticks(rotation=45)
plt.legend()

# Exibir os gráficos
plt.show()
```
