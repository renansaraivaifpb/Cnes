```python
import pandas as pd
```

```python
tabela_investigacao_de_obito =  pd.read_csv("https://raw.githubusercontent.com/renansaraivaifpb/analise_dados_saude_brasileira/main/dados/obitos_fetais/tabela_investigacao_de_obito.csv")
tabela_obito_por_parto =  pd.read_csv("https://raw.githubusercontent.com/renansaraivaifpb/analise_dados_saude_brasileira/main/dados/obitos_fetais/tabela_obito_por_parto.csv")
tabela_obito_p_ocorrencia_p_obito_relacao_parto_segundo_municipio = pd.read_csv("https://raw.githubusercontent.com/renansaraivaifpb/analise_dados_saude_brasileira/main/dados/obitos_fetais/tabela_obito_p_ocorrencia_p_obito_relacao_parto_segundo_nunicipio.csv")
tabela_obitos_p_correncia_por_peso_ao_nascer_segundo_municipio = pd.read_csv("https://raw.githubusercontent.com/renansaraivaifpb/analise_dados_saude_brasileira/main/dados/obitos_fetais/tabela_obitos_p_correncia_por_peso_ao_nascer_segundo_municipio.csv")
tabela_obitos_p_correncia_por_idade_mae_segundo_municipio = pd.read_csv("https://raw.githubusercontent.com/renansaraivaifpb/analise_dados_saude_brasileira/main/dados/obitos_fetais/tabela_obitos_p_correncia_por_idade_mae_segundo_municipio.csv")
tabela_obitos_p_ocorrencia_por_escolaridade_mae_segundo_municipio = pd.read_csv("https://raw.githubusercontent.com/renansaraivaifpb/analise_dados_saude_brasileira/main/dados/obitos_fetais/tabela_obitos_p_ocorrencia_por_escolaridade_mae_segundo_municipio.csv")
tabela_taxa_mortalidade_media_por_municipio = pd.read_csv("https://raw.githubusercontent.com/renansaraivaifpb/analise_dados_saude_brasileira/main/dados/obitos_fetais/tabela_taxa_mortalidade_media_por_municipio.csv")
```

```python
tabela_investigacao_de_obito = tabela_investigacao_de_obito.replace("-",0)
tabela_obito_por_parto = tabela_obito_por_parto.replace("-",0)
tabela_obito_p_ocorrencia_p_obito_relacao_parto_segundo_municipio = tabela_obito_p_ocorrencia_p_obito_relacao_parto_segundo_municipio.replace("-",0)
tabela_obitos_p_correncia_por_peso_ao_nascer_segundo_municipio = tabela_obitos_p_correncia_por_peso_ao_nascer_segundo_municipio.replace("-",0)
tabela_obitos_p_correncia_por_idade_mae_segundo_municipio = tabela_obitos_p_correncia_por_idade_mae_segundo_municipio.replace("-",0)
tabela_obitos_p_ocorrencia_por_escolaridade_mae_segundo_municipio = tabela_obitos_p_ocorrencia_por_escolaridade_mae_segundo_municipio.replace("-",0)
```

```python
tabela_investigacao_de_obito = tabela_investigacao_de_obito.iloc[:,:-1]
tabela_investigacao_de_obito = tabela_investigacao_de_obito.drop(0)

tabela_obito_por_parto = tabela_obito_por_parto.iloc[:,:-1]

tabela_obito_p_ocorrencia_p_obito_relacao_parto_segundo_municipio = tabela_obito_p_ocorrencia_p_obito_relacao_parto_segundo_municipio.drop(0)
tabela_obito_p_ocorrencia_p_obito_relacao_parto_segundo_municipio = tabela_obito_p_ocorrencia_p_obito_relacao_parto_segundo_municipio.iloc[:,:-2]

tabela_obitos_p_correncia_por_peso_ao_nascer_segundo_municipio = tabela_obitos_p_correncia_por_peso_ao_nascer_segundo_municipio.drop(0)
tabela_obitos_p_correncia_por_peso_ao_nascer_segundo_municipio = tabela_obitos_p_correncia_por_peso_ao_nascer_segundo_municipio.iloc[:,:-2]

tabela_obitos_p_correncia_por_idade_mae_segundo_municipio = tabela_obitos_p_correncia_por_idade_mae_segundo_municipio.drop(0)
tabela_obitos_p_correncia_por_idade_mae_segundo_municipio = tabela_obitos_p_correncia_por_idade_mae_segundo_municipio.iloc[:,:-2]

tabela_obitos_p_ocorrencia_por_escolaridade_mae_segundo_municipio = tabela_obitos_p_ocorrencia_por_escolaridade_mae_segundo_municipio.drop(0)
tabela_obitos_p_ocorrencia_por_escolaridade_mae_segundo_municipio = tabela_obitos_p_ocorrencia_por_escolaridade_mae_segundo_municipio.iloc[:,:-2]


```

```python
resultado = tabela_investigacao_de_obito.merge(tabela_obito_por_parto, on='Município', how='right').fillna(0)
resultado = resultado.merge(tabela_obito_p_ocorrencia_p_obito_relacao_parto_segundo_municipio, on='Município', how='right').fillna(0)
resultado = resultado.merge(tabela_obitos_p_correncia_por_peso_ao_nascer_segundo_municipio, on='Município', how='right').fillna(0)
resultado = resultado.merge(tabela_obitos_p_correncia_por_idade_mae_segundo_municipio, on='Município', how='right').fillna(0)
resultado = resultado.merge(tabela_obitos_p_ocorrencia_por_escolaridade_mae_segundo_municipio, on='Município', how='right').fillna(0)
resultado = resultado.merge(tabela_taxa_mortalidade_media_por_municipio, on='Município', how='right').fillna(0)
resultado
```

```python
resultado.iloc[:, 1:] = resultado.iloc[:, 1:].astype('float')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Seleciona as colunas que deseja incluir na matriz de correlação
subset = resultado.iloc[:, 1:]

# Calcula a matriz de correlação
correlation_matrix = subset.corr()

# Cria um mapa de calor (heatmap) da matriz de correlação
plt.figure(figsize=(24, 18))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlação")
plt.show()
```
