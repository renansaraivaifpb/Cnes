```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponha que você tenha um DataFrame chamado df com as colunas CNES, Estabelecimento, Municipios, Existentes, SUS e Estado

# Crie um dicionário que mapeia cada estado para sua região correspondente
regioes = {
    'ACRE': 'Oeste',
    'ALAGOAS': 'Nordeste',
    'AMAPA': 'Norte',
    'AMAZONAS': 'Norte',
    'BAHIA': 'Nordeste',
    'CEARA': 'Nordeste',
    'DISTRITO FEDERAL': 'Leste',
    'ESPIRITO SANTO': 'Leste',
    'GOIAS': 'Leste',
    'MARANHAO': 'Nordeste',
    'MATO GROSSO': 'Oeste',
    'MATO GROSSO DO SUL': 'Oeste',
    'MINAS GERAIS': 'Leste',
    'PARA': 'Norte',
    'PARAIBA': 'Nordeste',
    'PARANA': 'Sul',
    'PERNAMBUCO': 'Nordeste',
    'PIAUI': 'Nordeste',
    'RIO DE JANEIRO': 'Leste',
    'RIO GRANDE DO NORTE': 'Nordeste',
    'RIO GRANDE DO SUL': 'Sul',
    'RONDONIA': 'Oeste',
    'SANTA CATARINA': 'Sul',
    'SAO PAULO': 'Leste',
    'SERGIPE': 'Nordeste',
    'TOCANTINS': 'Oeste'
}

# Adicione uma nova coluna 'Regiao' ao DataFrame com base no mapeamento
df['Regiao'] = df['Estado'].map(regioes)

# Crie subplots para cada região
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Itere sobre as regiões e crie um plot para cada uma
for i, regiao in enumerate(['Nordeste', 'Sul', 'Leste', 'Oeste']):
    ax = axs[i // 2, i % 2]
    soma_por_regiao = df[df['Regiao'] == regiao].groupby('Estado')['Existentes'].sum()
    soma_por_regiao.plot(kind='bar', ax=ax)
    ax.set_title(f'UTI NEONATAL - TIPO I Existentes - {regiao}')
    ax.set_xlabel('Estado')
    ax.set_ylabel('Soma dos Valores Existentes')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Lista dos nomes dos arquivos CSV
arquivos_csv = [
    "UCI PEDIATRICA - NEONATAL CONVENCIONAL.csv",
    "UTI NEONATAL - TIPO I.csv",
    "UTI NEONATAL - TIPO II.csv",
    "UTI NEONATAL - TIPO III.csv",
    "UTI PEDIATRICA - TIPO I.csv",
    "UTI PEDIATRICA - TIPO II.csv",
    "UTI PEDIATRICA - TIPO III.csv"
]

# Crie subplots para cada arquivo CSV
fig, axs = plt.subplots(3, 3, figsize=(24, 24))

# Itere sobre os arquivos CSV
for i, arquivo in enumerate(arquivos_csv):
    # Verifique se o arquivo existe
    if os.path.exists(arquivo):
        # Leia o arquivo CSV em um DataFrame
        df = pd.read_csv(arquivo)
        
        # Crie um mapeamento de estado para região (como no código original)
        regioes = {
            'ACRE': 'Oeste',
            'ALAGOAS': 'Nordeste',
            'AMAPA': 'Norte',
            'AMAZONAS': 'Norte',
            'BAHIA': 'Nordeste',
            'CEARA': 'Nordeste',
            'DISTRITO FEDERAL': 'Leste',
            'ESPIRITO SANTO': 'Leste',
            'GOIAS': 'Leste',
            'MARANHAO': 'Nordeste',
            'MATO GROSSO': 'Oeste',
            'MATO GROSSO DO SUL': 'Oeste',
            'MINAS GERAIS': 'Leste',
            'PARA': 'Norte',
            'PARAIBA': 'Nordeste',
            'PARANA': 'Sul',
            'PERNAMBUCO': 'Nordeste',
            'PIAUI': 'Nordeste',
            'RIO DE JANEIRO': 'Leste',
            'RIO GRANDE DO NORTE': 'Nordeste',
            'RIO GRANDE DO SUL': 'Sul',
            'RONDONIA': 'Oeste',
            'SANTA CATARINA': 'Sul',
            'SAO PAULO': 'Leste',
            'SERGIPE': 'Nordeste',
            'TOCANTINS': 'Oeste'
        }
        
        # Adicione uma nova coluna 'Regiao' ao DataFrame com base no mapeamento
        df['Regiao'] = df['Estado'].map(regioes)
        
        # Selecione o subplot apropriado
        ax = axs[i // 3, i % 3]
        
        # Agrupe e plote os dados
        soma_por_regiao = df.groupby('Estado')['Existentes'].sum()
        soma_por_regiao.plot(kind='bar', ax=ax)
        ax.set_title(f'{arquivo} - Quantidade de Leitos existentes por Estado')
        ax.set_xlabel('Estado')
        ax.set_ylabel('Somatório Existentes')
        
    else:
        print(f"Arquivo {arquivo} não encontrado.")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponha que você tenha três DataFrames: df_uci_pediatrica, df_uti_neonatal e df_uti_pediatrica,
# cada um contendo as colunas CNES, Existentes e UF.
df_uci_pediatrica = pd.read_csv("UCI PEDIATRICA - NEONATAL CONVENCIONAL.csv", sep=',',encoding='latin-1')

df_uti_neonatal = pd.read_csv("UTI NEONATAL - TIPO I.csv", sep=',',encoding='latin-1')

df_uti_pediatrica = pd.read_csv("UTI PEDIATRICA - TIPO I.csv", sep=',',encoding='latin-1')


# Calcule a média da quantidade de UCIs Pediátricas por UF
media_uci_pediatrica = df_uci_pediatrica.groupby('Estado')['Existentes'].mean()

# Calcule a média da quantidade de UTIs Neonatais por UF
media_uti_neonatal = df_uti_neonatal.groupby('Estado')['Existentes'].mean()

# Calcule a média da quantidade de UTIs Pediátricas por UF
media_uti_pediatrica = df_uti_pediatrica.groupby('Estado')['Existentes'].mean()

# Crie um gráfico comparativo
plt.figure(figsize=(12, 6))
plt.bar(media_uci_pediatrica.index, media_uci_pediatrica, label='UCI Pediátrica - NEONATAL CONVENCIONAL', alpha=0.72)
plt.bar(media_uti_neonatal.index, media_uti_neonatal, label='UTI Neonatal - TIPO I', alpha=0.7)
plt.bar(media_uti_pediatrica.index, media_uti_pediatrica, label='UTI Pediátrica - TIPO I', alpha=0.7)
plt.xlabel('UF')
plt.ylabel('Média da Quantidade de Unidades')
plt.title('Comparação da Média da Quantidade de UCIs e UTIs por UF')
plt.legend()
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
